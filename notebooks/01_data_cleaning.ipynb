## Feature Engineering

```python
df['tenure_bucket'] = pd.cut(df['tenure'], 
                             bins=[-1,6,12,24,48,100], 
                             labels=['0-6','7-12','13-24','25-48','49+'])
```

**'No','No internet service','No phone service'**
These are also present in the data entries with No and Yes

```python
service_cols = ['PhoneService','InternetService','OnlineSecurity','OnlineBackup',
                'DeviceProtection','TechSupport','StreamingTV','StreamingMovies']

def count_services(row):
    count = 0
    for col in service_cols:
        if row[col] not in ['No', 'No phone service', 'No internet service']:
            count += 1
    return count

df['num_services'] = df.apply(count_services, axis=1)
```

```python
df.head()
```

```python
df['avg_monthly_charge'] = df['TotalCharges'] / df['tenure'].replace(0,1)
```
